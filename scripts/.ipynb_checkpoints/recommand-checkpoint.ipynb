```python
#aggration user action data from json file
import os
import json
def load_json(fname):
    with open(fname) as json_file:
        return json.load(json_file)
        
def get_json_file_list(data_dir):
    '''get all json file list from data_dir'''
    res = []
    for dir in os.listdir(data_dir):
        base = os.path.join(data_dir, dir)
        for json_file in os.listdir(base):
            f = os.path.join(base, json_file)
            yield f

def merge_dict(a, b):
    '''merge two dictionary to one'''
    res = a
    for gid, uid_dict in b.items():
        if not res.has_key(gid):
            res[gid] = uid_dict
            continue
        for uid, op_dict in uid_dict.items():
            if not res[gid].has_key(uid):
                res[gid][uid] = op_dict
                continue
            for op, op_val in op_dict.items():
                if not res[gid][uid].has_key(op):
                    res[gid][uid][op] = op_val
                    continue
                res[gid][uid][op] += op_val
    return res
        
    
def combine_json_data(data_dir, ofile):
    '''combine all data of every day from data_dir'''
    print 'begin to combine json data...'
    
    all = dict()
    for json_file in get_json_file_list(data_dir):
        src = load_json(json_file)
        all = merge_dict(all, src)
        
    print 'done to combine json data...'
    
    with open(ofile, 'w') as f:
        json.dump(all, f)
    return all

all = combine_json_data('../json-res', '../data/all_user_data.json')
              
```

```python
#anaylize user action data
%matplotlib inline
import numpy as np
import pylab as plt
import matplotlib.pyplot as pyplot # 类似matlab绘图命令包
import matplotlib.font_manager
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

all_data_path = '../data/all_user_data.json'
def analize_group_popular(all_data):
    popular = dict()
    for gid, uid_dict in all_data.items():
        num_user = len(uid_dict)
        if not popular.has_key(num_user):
            popular[num_user] = 0
        popular[num_user] += 1
    
    return popular

def func_2nd(x):
    return (x-200)/5 + 200

def func_3nd(x):
    return (x-1000)/50 + func_2nd(1000)

def plot_result(popular):
    x = popular.keys()
    y = popular.values()
    for i in range(len(x)):
        if x[i]>= 200 and x[i] < 1000:
            x[i] = func_2nd(x[i])
        elif x[i] >= 1000 and x[i] < 7000:
            x[i] = func_3nd(x[i])
    linewidth = 3.5
    spineswidth = 3.5
    gridlinewidth = 0.5
    fontsize = 10
    pad = 10
    isGridShow = True
    #filename = raw_input()
    dpi = 1000
    font = matplotlib.font_manager.FontProperties(family='times new roman', style='normal', size=fontsize, weight='normal', stretch='normal')
    fig, ax = pyplot.subplots()
    ax.scatter(x,y)
    ax.set_xlabel('popular', fontproperties=font)
    ax.set_ylabel('group number', fontproperties=font)
    ax.set_xlim(-20, func_3nd(9000))
    ax.set_ylim(0, 40)
    # 设置x，y轴的ticks和ticklabels，完全自由控制
    ax.set_xticks([1,10,100,200,300,1000])
    ax.set_xticklabels( [1,10,100,300,1000,10000])
    ax.xaxis.set_minor_locator(MultipleLocator(20))
    ax.set_yticklabels( [0,10,40,80,130,200,500,1000,10000])
    #这个可以控制ticks的长短粗细，上下左右轴显示与否，牛逼不？
    ax.tick_params('both', length=5, width=0.8, which='major', pad=pad)
    ax.tick_params('both', length=4, width=0.8, which='minor', pad=pad)
    
        #这个可以控制ticklabels离坐标的距离等，牛逼！
    ax.tick_params('both',direction='in', which='both', pad=pad, bottom = 'on', top = 'on', left = 'on', right = 'on', labelcolor='black')
    fig.show()

all_data = load_json('../data/all_user_data.json')
plot_result(analize_group_popular(all_data))

#结果显示，群组的流行度分布符合长尾分布
```

```python
#挑选训练集和测试集

import random

def make_train_data(all_data):
    data = []
    for gid, uid_dict in all_data.items():
        for uid, op_dict in uid_dict.items():
            online =  op_dict.get('online')
            tmp = (uid, gid,online)
            data.append(tmp)
    return data

def splitData(data, M, K, seed):
    '''
        每次选取不同的K(0 <= K <= M-1)和相同的随机数种子seed，进行M次试验就可以得到M个不同的训练集和测试集
    '''
    train = []
    test = []
    random.seed(seed)
   
    for uid, gid, online in data:
        if online is None:
            online = 0
        if random.randint(0, M) == K:
            test.append([uid, gid, online])
        else:
            train.append([uid, gid, online])
   
    with open('../data/train.txt', 'w') as f:
        for a in train:
            uid = a[0]
            gid = a[1]
            online = a[2]
            f.writelines(str(uid) + ',' + str(gid) + ',' + str(online) + '\n')
    
    with open('../data/test.txt', 'w') as f:
        for a in test:
            uid = a[0]
            gid = a[1]
            online = a[2]
            f.writelines(str(uid) + ',' + str(gid) + ',' + str(online) + '\n')
    return train, test

M = 4
K = 2
seed = 7
train, test = splitData(make_train_data(all), M, K, seed)
```

```python
#量化评测指标
def Recall(train, test, N):
    hit = 0
    sumd = 0
    for user in train.keys():
        if user not in test:
            continue
        tu = test[user]
        rank = GetRecommendation(user, N)
        for group, pui in rank.items():#pui 评分
            if group in tu:
                hit += 1
        sumd += len(tu)
    return float(hit) / sumd

def Precision(train, test, N):
    hit = 0
    sumd = 0
    for user in train.keys():
        if user not in test:
            continue
        tu = test[user]
        rank = GetRecommendation(user, N)
        for gid, pui in rank.items():
            if gid in tu:
                hit += 1
        sumd += N
    return float(hit) / sumd

def Coverage(train, test, N):
    recommend_groups = set()
    all_gid = set()
    for user in train.keys():
        for gid in train[user].keys():
            all_gid.add(gid)
        rank = GetRecommendation(user, N)
        for gid, pui in rank.items():
            recommend_groups.add(gid)
    return float(len(recommend_groups)) / len(all_gid)

def Popularity(train, test, N):
    group_popularity = dict()
    for user, groups in train.items():
        for group in groups.keys():
            if group not in group_popularity:
                group_popularity[group] = 0
            group_popularity[group] += 1
    ret = 0
    n = 0
    for user in train.keys():
        rank = GetRecommendation(user, N)
        for group, pui in rank.items():
            ret += math.log(1 + group_popularity[group])
            n += 1
    ret /= n * 1.0
    return ret
```

```python
import math
import operator
import string
def getReverList(train_file):
    reverse = dict()
    with open(train_file) as f:
        for line in f:
            a = line.strip().split(',')
            uid = a[0]
            gid = a[1]
            online = string.atof(a[2])
            if uid not in reverse:
                reverse[uid] = dict()
            if gid not in reverse[uid]:
                reverse[uid][gid] = 0
            reverse[uid][gid] = online
    return reverse
def GroupSimilarity(train):
    '''
        train 用户-群组倒排表
    '''
    C = dict()  #贡献矩阵，C[i][j]表示群组i和群组j中共同的用户数
    N = dict()  #{gid: user_number_in_this_group}
    for u, groups in train.items():
        for i in groups:
            if i not in N:
                N[i] = 0
            if i not in C:
                C[i] = dict()
            N[i] += 1
            for j in groups:
                if i == j:
                    continue
                if j not in C[i]:
                    C[i][j] = 0
                C[i][j] += 1
    W = dict() #相似度矩阵
    for i, related_groups in C.items():
        if i not in W:
            W[i] = dict()
        for j, cij in related_groups.items():
            W[i][j] = cij / math.sqrt(N[i] * N[j])
    return W

def Recommendation(train, user_id, W, K):
    rank = dict()
    ru = train[user_id] #用户加入的群组集合
    for i, online in ru.items():
        for j, wj in sorted(W[i].items(), key=operator.itemgetter(1), reverse=True)[0:K]:
            if j in ru:
                continue
            if j not in rank:
                rank[j] = 0
            rank[j] += float(online) * wj
    return rank

```

```python
train_revers = getReverList('../data/train.txt')
test_revers = getReverList('../data/test.txt')
W = GroupSimilarity(train_revers)                                                                                                                  
def GetRecommendation(user, N):  
    return Recommendation(train_revers, user, W, N)

for i in range(5, 25, 5):
    recall = Recall(train_revers, test_revers, i)
    precision = Precision(train_revers, test_revers,i)
    coverge = Coverage(train_revers, test_revers, i)
    popularity = Popularity(train_revers, test_revers, i)
    print '当N = ' + str(i) + '时， Precision = ' + str(precision) + ', Recall = ' + str(recall) + ', Coverge = ' + str(coverge) + ', Popularity= ' + str(popularity)
```
